### Agora Ã© preciso que eu transforme essas colunas em linhas, de forma que seja possÃ­vel analiza-lÃ¡s

O passo a passo Ã© o seguinte:

1. Transformar as colunas em linhas ( terei de pesquisar como fazer )

2. Guardar isso em uma nova variÃ¡vel

3. Fazer isso para os 2 dataframes

4. Aplicar transformaÃ§Ãµes extras necessÃ¡rias

```python
# Para transformar as colunas em linhas em python usarei o mÃ©todo .melt do pandas .
df_pop_19_pivoted = df_pop_19.melt( id_vars="NAME",
                                    var_name="YEAR",
                                    value_name="POPESTIMATE" )
                                   
```

```python
df_pop_19_pivoted
```

```python
# O mesmo para 2022
df_pop_22_pivoted = df_pop_22.melt( id_vars="NAME",
                                   var_name="YEAR",
                                   value_name="POPESTIMATE"  )
```

```python
# Checando o formato das duas tabelas, ambas tÃªm 3 colunas, e agora Ã© possÃ­vel juntar as 2 tabelas em uma sÃ³, facilitando qualquer anÃ¡lise
print(df_pop_19_pivoted.shape)
print(df_pop_22_pivoted.shape)
```

```python
# Como sÃ³ me interessa dados para os Estados Unidos
df_pop_22_pivoted_US = df_pop_22_pivoted[df_pop_22_pivoted["NAME"] == "United States"]
```

```python
df_pop_19_pivoted_US = df_pop_19_pivoted[df_pop_19_pivoted["NAME"] == "United States"]
```

```python
# Printando dessa forma pude me lembrar que precisarei cuidar para que os cabeÃ§alhos nÃ£o sejam incluÃ­dos no momento de juntar as tabelas
print( df_pop_22_pivoted_US.head() )
print( df_pop_19_pivoted_US.head() )
```

```python
Sensus_US = pd.concat([ df_pop_19_pivoted_US, df_pop_22_pivoted_US ], ignore_index=True)
```

```python
Sensus_US.head()
```

```python
Sensus_US['YEAR'] = Sensus_US['YEAR'].str.extract('(\d+)').astype(int) # Pegando apenas o ano de cada Linha.
```

```python
Sensus_US
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))  # Definindo o tamanho da figura
plt.plot( Sensus_US['YEAR'], Sensus_US['POPESTIMATE'], marker='o')  # Criando o grÃ¡fico de linha com marcadores em cada ponto

# Adicionando tÃ­tulo e rÃ³tulos aos eixos
plt.title('PopulaÃ§Ã£o dos Estados Unidos ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('PopulaÃ§Ã£o Estimada')
plt.legend('US')
# Mostrando a grade para melhor leitura
plt.grid(True)

# Exibindo o grÃ¡fico
plt.show()
```

```python
# entre 2019 e 2022 houve uma mudanÃ§a na tendÃªncia de crescimento da populaÃ§Ã£o, seria interessante saber qual foi a mudanÃ§a em %
```

#### Entre 2019 e 2022 houve uma mudanÃ§a na tendÃªncia de crescimento da populaÃ§Ã£o,

```python
Sensus_US['Yearly_Change'] = Sensus_US['POPESTIMATE'].pct_change() * 100
Sensus_US
```
