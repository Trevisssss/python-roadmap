```python
import pandas as pd
import requests
import json
```

<b style = "font-size: 28px; color: #FFD166" > PARA ESTE EXEMPLO Ã‰ UTILIZADO A API DO POKEMON </b>

```python
url = "https://pokeapi.co/api/v2/pokemon/"
pokemon_list = []

while url != None: #CRIADO PARA MONTAR A PAGINAÃ‡Ã‚O ( Percorrer outras pÃ¡ginas )
    
    payload = {}
    headers = {}
    response = json.loads( requests.request("GET", url, headers=headers, data=payload).text )
    url = response['next'] ##Esta Chave retorna o link para a prÃ³xima pÃ¡gina. ( ESPECIFICO DESTA API )
    
    for item in response['results']:
        pokemon_name = item['name']
        url_pokemon = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
        response_pokemon = json.loads( requests.request("GET", url_pokemon, headers=headers, data=payload).text )
        
        infos = {
            "name": pokemon_name,
            "id": response_pokemon['id'],
            "heigth": response_pokemon['height'],
            "weight": response_pokemon['weight'],
            "is_default": response_pokemon['is_default']
        }
        
        pokemon_list.append(infos)
        print(response_pokemon['id'])
        
pokemon_list
```

```python
pokemon_dataframe = pd.DataFrame(pokemon_list)
```
